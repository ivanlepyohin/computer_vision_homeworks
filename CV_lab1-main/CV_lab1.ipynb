```python
img_bgr = cv2.imread('/content/drive/MyDrive/CV_lab1-main/cat.png')
img_bgr2hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

img_rgb2hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)

#cv2_imshow(img_rgb2hsv)
cv2_imshow(img_bgr2hsv)

print(f"Размер фотографии с котиком = {np.array(img_bgr).shape}")
(b, g, r) = img_rgb2hsv[70, 70]
print("Красный: {}, Зелёный: {}, Синий: {}".format(r, g, b))
print(img_rgb2hsv[70, 70])
print(img_rgb.dtype)
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

hsv_arr = np.zeros_like(img_rgb, dtype=np.uint8)

for i in range(img_rgb.shape[0]):
  for j in range(img_rgb.shape[1]):
    r,g,b = img_rgb[i][j]
    r, g, b = r / 255.0, g / 255.0, b / 255.0

    v = max(r, g, b)
    temp = min(r, g, b)

    if v == 0:
        s = 0
    else:
        s = (v - temp) / v

    if r == g == b:
       h = 0
    elif v == r:
      h = 60 * (g - b) / (v - temp)
    elif v == g:
        h = 120 + 60 * (b - r) / (v - temp)
    elif v == b:
        h = 240 + 60 * (r - g) / (v-temp)


    if h < 0:
        h = h + 360

    h = h/2
    s = 255*s
    v = 255*v

    hsv_arr[i, j] = [h, s, v]

cv2_imshow(hsv_arr)
(b, g, r) = hsv_arr[70, 70]
print(b, g, r)
```
